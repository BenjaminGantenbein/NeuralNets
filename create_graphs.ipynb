```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

# Read in CSV file
with open("graph_data_transformer.csv", "r") as f:
    csv_reader = csv.reader(f)
    next(csv_reader)  # skip header row
    train_loss = []
    val_loss = []
    val_acc = []
    for line in csv_reader:
        if not line:  # check if line is empty
            break  # break out of the loop if the line is empty
        train_loss.append(float(line[0]))
        val_loss.append(float(line[1]))
        val_acc.append(float(line[2]))

# Set plot style
plt.style.use('ggplot')

# Create a plot of the loss
plt.figure(figsize=(10,6))
plt.plot(train_loss, color='tab:blue', label='train loss')
plt.plot(val_loss, color='tab:orange', label='val loss')
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.title('Training and Validation Loss', fontsize=16)
plt.show()

# Create a plot of the accuracy
plt.figure(figsize=(10,6))
plt.plot(val_acc, color='tab:green', label='val acc')
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.title('Validation Accuracy', fontsize=16)
plt.show()
```
